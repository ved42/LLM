```python
from dotenv import load_dotenv
from langchain.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain_community.document_loaders import PyPDFLoader
import os
```

```python
loader = PyPDFLoader('apjspeech.pdf')
docs=loader.load_and_split()
docs
```

```python
len(docs)
```

```python
## Load the enviorment variable
load_dotenv()
api_key = os.getenv("GROQ_API_KEY")
```

```python
llm = ChatGroq(groq_api_key=api_key,model="Gemma-7b-It")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
final_doc = RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=200).split_documents(docs)
final_doc
```

```python
len(final_doc)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
## Creating a prompt
promptt=""" 
You are a helpful assistant.Based on given description.Generate a summary with a meaningful title.
Description:{text}

"""
##Template
summary_temp = PromptTemplate(input_variables=['text'],template=promptt)
```

```python
final_prompt = ''' 
Provide the final summary of the entire speech with these important points.
Add a Motivation Title,Start the precise summary with an introduction and provide the summary in number 
points for the speech.
Speech:{text}

'''
final_prompt_template = PromptTemplate(input_variables=['text'],template=final_prompt)
```

```python
summary_chain = load_summarize_chain(llm=llm,
                                     chain_type='map_reduce',
                                     map_prompt=summary_temp,
                                     combine_prompt=final_prompt_template,
                                     verbose=True)

output = summary_chain.run(final_doc)
```

### Email Reply Assistant 

```python
from dotenv import load_dotenv
from langchain.prompts import PromptTemplate
from langchain_groq import ChatGroq
import os

## Load the enviorment variable
load_dotenv()
api_key = os.getenv("GROQ_API_KEY")

llm = ChatGroq(groq_api_key=api_key,model="Gemma-7b-It")

# Email prompt template with sender and receiver names
email_Prompt = '''
Write a professional and polite email reply from {sender_name} to {receiver_name} based on the following details:

Subject: {subject}

Email Message:
{message}

Please ensure the reply addresses the points mentioned in the email, and provide a clear, courteous, and formal tone. The email should conclude with an appropriate closing and sign-off.

Reply:
'''


email_prompt_temp = PromptTemplate(input_variables=['sender_name','receiver_name','subject','message'],template=email_Prompt)

sender_name='vedant'
receiver_name='kalpak'
subject = "Meeting Schedule for Q4 Project"
message = """
Hi, I wanted to check if you're available for a meeting next week to discuss the Q4 project milestones. 
Please let me know your availability, and if you have any specific agenda items you'd like to cover.
"""
filled_prompt=email_prompt_temp.format(sender_name=sender_name,receiver_name=receiver_name,subject=subject,message=message)

response =llm.invoke(filled_prompt)

genrated_response = response.content

print(genrated_response)
```
